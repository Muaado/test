## Modeling Start
This section partitions the dataset as is, and builds a decision tree based adaboost model which serves as a base model.

```python
# partition the data
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

# Begin Sampling Strategies

# 1. Training Sample

```python
# converts training set to ndarray
train_x = x_train.values
train_y = y_train.values
```

```python
# Preparing Training Set
train_x_h2 = h2o.H2OFrame(train_x)
train_y_h2 = h2o.H2OFrame(train_y)
train_y_h2.columns = ['Class']

train_h2 = train_x_h2.cbind(train_y_h2)

train_h2['Class'] = train_h2['Class'].asfactor()
train_h2.columns = list(j for c,j in enumerate(data.columns))
```

```python
train_h2_1 = train_h2.as_data_frame()
%store train_h2_1
```

# 2. RUS - Random UnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=21333)
rus_x, rus_y = rus.fit_resample(train_x, train_y)
print('Random undersampling {}'.format(Counter(rus_y)))
```

```python
# Preparing Training Set
rus_x_h2 = h2o.H2OFrame(rus_x)
rus_y_h2 = h2o.H2OFrame(rus_y)
rus_y_h2.columns = ['Class']

rus_h2 = rus_x_h2.cbind(rus_y_h2)

rus_h2['Class'] = rus_h2['Class'].asfactor()
rus_h2.columns = list(j for c,j in enumerate(data.columns))
```

```python
rus_h2_1 = rus_h2.as_data_frame()
%store rus_h2_1
```

# 3. SMOTE - Synthetic Minority Oversampling

```python
import smote_variants as smoter
smotes = smoter.SMOTE()
smote_x, smote_y = smotes.sample(train_x, train_y)
print('SMOTE oversampling {}'.format(Counter(smote_y)))
```

```python
# Preparing Training Set
smote_x_h2 = h2o.H2OFrame(smote_x)
smote_y_h2 = h2o.H2OFrame(smote_y)
smote_y_h2.columns = ['Class']

smote_h2 = smote_x_h2.cbind(smote_y_h2)

smote_h2['Class'] = smote_h2['Class'].asfactor()
smote_h2.columns = list(j for c,j in enumerate(data.columns))
```

```python
smote_h2_1 = smote_h2.as_data_frame()
%store smote_h2_1
```

# 4. DBSMOTE - DBSynthetic Minority Oversampling

```python
dbsmotes = smoter.DBSMOTE(proportion=1, n_jobs=10)
dbsmote_x, dbsmote_y = dbsmotes.sample(train_x, train_y)
print('DBSMOTE oversampling {}'.format(Counter(dbsmote_y)))
```

```python
# Preparing Training Set
dbsmote_x_h2 = h2o.H2OFrame(dbsmote_x)
dbsmote_y_h2 = h2o.H2OFrame(dbsmote_y)
dbsmote_y_h2.columns = ['Class']

dbsmote_h2 = dbsmote_x_h2.cbind(dbsmote_y_h2)

dbsmote_h2['Class'] = dbsmote_h2['Class'].asfactor()

dbsmote_h2.columns = list(j for c,j in enumerate(data.columns))
```

```python
dbsmote_h2_1 = dbsmote_h2.as_data_frame()
%store dbsmote_h2_1
```

# 5. SMOTEENN - Hybrid - Tomeklinks and SMOTE

```python
smote_eenn = smoter.SMOTE_ENN(proportion=1, n_jobs=10)
smoteenn_x, smoteenn_y = smote_eenn.sample(train_x, train_y)
print('SMOTE_ENN hybrid sampling {}'.format(Counter(smoteenn_y)))
```

```python
# Preparing Training Set
smoteenn_x_h2 = h2o.H2OFrame(smoteenn_x)
smoteenn_y_h2 = h2o.H2OFrame(smoteenn_y)
smoteenn_y_h2.columns = ['Class']

smoteenn_h2 = smoteenn_x_h2.cbind(smoteenn_y_h2)

smoteenn_h2['Class'] = smoteenn_h2['Class'].asfactor()

smoteenn_h2.columns = list(j for c,j in enumerate(data.columns))
```

```python
smoteenn_h2_1 = smoteenn_h2.as_data_frame()
%store smoteenn_h2_1
```

```python
# Preparing Holdoutset
testset = pd.merge(x_test, y_test , right_index = True, left_index = True, how='left')
testset_h2 = h2o.H2OFrame(testset)
testset_h2['Class'] = testset_h2['Class'].asfactor()
```
